```python
def make_diag_nan(A):
    A[np.eye(A.shape[0],dtype=bool)] = np.nan
    return A

save_path = "similarity_evaluation_portfolio1.csv"
extracted_mf = pd.read_csv(save_path, index_col="data_id").sort_index()
extracted_mf.loc[extracted_mf.index.isin(extracted_mfe.index)]

# make mfe work
extracted_mf = extracted_mf.loc[extracted_mf.index.isin(extracted_mfe.index)]

extracted_mf_rank = extracted_mf.drop(["name","Unnamed: 0"], axis = 1).rank(axis=1,ascending=False).sort_index()
extracted_mf_np = extracted_mf_rank.to_numpy()
distances = -np.ones((extracted_mf_np.shape[0],)*2)
for row1 in range(extracted_mf_np.shape[0]):
    for row2 in range(extracted_mf_np.shape[0]):
        # print( (extracted_mf_np[row2], extracted_mf_np[row2])[0])
        distances[row1, row2] = kendalltau(extracted_mf_np[row1], extracted_mf_np[row2])[0]
distances = make_diag_nan(distances)
GT_ranking = pd.DataFrame(make_diag_nan(distances), index=extracted_mf_rank.index, columns=extracted_mf_rank.index).rank(1, ascending=False)
display(GT_ranking)
```

```python
from evaluate_similarity import create_ranking
names = ["d2v", "mfe", "pfn"]
save_path = ["../extracted_MF/OpenML-CC18_d2v.csv", "../extracted_MF/OpenML-CC18_mfe.csv", "../extracted_MF/OpenML-CC18_pfn.csv"]
ranks     = [pd.DataFrame(index=GT_ranking.index, columns=GT_ranking.index)]*3
res = {}
for name, path, r in zip(names,save_path, ranks):
    for did in GT_ranking.index:
        ranking = create_ranking(did, path)
        # ranking = ranking.set_index("did")
        r.loc[did,:] = ranking.set_index("did").T.loc["Cosine Similarity"]
        # break
    r = r.rank(ascending= False, axis=1)
    res[name] = r

# save_path = "../extracted_MF/OpenML-CC18_mfe.csv"
# mfe_r = pd.DataFrame(index=GT_ranking.index, columns=GT_ranking.index)
# for did in GT_ranking.index:
#     ranking = create_ranking(did, save_path)
#     # ranking = ranking.set_index("did")
#     mfe_r.loc[did,:] = ranking.set_index("did").T.loc["Cosine Similarity"]
#     # break
# mfe_r = mfe_r.rank(ascending= False, axis=1)

# save_path = "../extracted_MF/OpenML-CC18_pfn.csv"
# pfn_r = pd.DataFrame(index=GT_ranking.index, columns=GT_ranking.index)
# for did in GT_ranking.index:
#     ranking = create_ranking(did, save_path)
#     # ranking = ranking.set_index("did")
#     pfn_r.loc[did,:] = ranking.set_index("did").T.loc["Cosine Similarity"]
#     # break
# pfn_r = pfn_r.rank(ascending= False, axis=1)
```
